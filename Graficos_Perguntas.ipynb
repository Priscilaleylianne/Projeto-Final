```python
# Importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np # Para cálculos numéricos e geração de dados

# Configurações para melhorar a visualização dos gráficos
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (10, 7) # Tamanho da figura ajustado para pie chart
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']

print("Bibliotecas importadas e configurações de plotagem aplicadas.")

# --- Dados Simulados para o Gráfico de Pizza ---
# Vamos simular uma distribuição de produtos reordenados e não reordenados
np.random.seed(42) # Para reprodutibilidade

# Simular um grande número de itens de pedido com status 'reordered'
num_items = 10000
data_reorder = pd.DataFrame({
    'reordered': np.random.choice([0, 1], size=num_items, p=[0.35, 0.65]) # 35% não reordenado, 65% reordenado
})

# Contar a ocorrência de cada status
reorder_counts = data_reorder['reordered'].value_counts()

# Mapear os valores numéricos para rótulos significativos
labels = ['Não Recomprado', 'Recomprado']
sizes = [reorder_counts[0], reorder_counts[1]]
colors = ['#FF9999', '#66B2FF'] # Cores suaves para os setores

# --- Gráfico de Pizza: Proporção de Produtos Recomprados vs. Não Recomprados ---
plt.figure(figsize=(8, 8)) # Tamanho quadrado para o gráfico de pizza

# Criar o gráfico de pizza
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90,
        pctdistance=0.85, wedgeprops=dict(width=0.4)) # autopct para formatar percentual, wedgeprops para anel
plt.title('Proporção Geral de Produtos Recomprados vs. Não Recomprados', pad=20)
plt.axis('equal') # Garante que o círculo seja desenhado como um círculo.

# Adicionar um círculo central para um efeito de "donut"
circle = plt.Circle((0,0), 0.60, fc='white')
fig = plt.gcf()
fig.gca().add_artist(circle)

plt.tight_layout()
plt.savefig('proporcao_recompra_geral.png')
plt.show()
plt.close()
print("Gráfico 'proporcao_recompra_geral.png' gerado.")
```

```python
# -*- coding: utf-8 -*-
"""
Demonstração de Atributos Derivados (Engenharia de Features)
Projeto: Análise Exploratória de Dados do Instacart
"""

# ----------------------------------------------------------------------------
# Instalação e Importação de Bibliotecas
# ----------------------------------------------------------------------------

# Não é necessário instalar bibliotecas adicionais para este script,
# pois pandas, numpy e matplotlib já vêm pré-instalados no Google Colab.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Bibliotecas importadas com sucesso!")

# ----------------------------------------------------------------------------
# 1. Simulação dos Dados Brutos do Instacart
#    (Substituir pela leitura dos CSVs reais no seu projeto)
# ----------------------------------------------------------------------------

print("\n--- 1. Simulando Dados Brutos ---")

# Simular a tabela 'orders'
np.random.seed(42)
num_users = 100
num_orders = 1000
orders_data = {
    'order_id': range(1, num_orders + 1),
    'user_id': np.random.randint(1, num_users + 1, num_orders),
    'eval_set': np.random.choice(['prior', 'train'], num_orders, p=[0.8, 0.2]),
    'order_number': np.zeros(num_orders, dtype=int), # Será preenchido
    'order_dow': np.random.randint(0, 7, num_orders),
    'order_hour_of_day': np.random.randint(0, 24, num_orders),
    'days_since_prior_order': np.random.randint(1, 31, num_orders)
}
df_orders = pd.DataFrame(orders_data)

# Preencher order_number sequencialmente por user_id
df_orders['order_number'] = df_orders.groupby('user_id')['order_id'].rank(method='first').astype(int)
df_orders.loc[df_orders['order_number'] == 1, 'days_since_prior_order'] = np.nan # Primeiras ordens não tem prior

print("df_orders (simulado):")
print(df_orders.head())

# Simular a tabela 'products'
num_products = 500
products_data = {
    'product_id': range(1, num_products + 1),
    'product_name': [f'Produto_{i}' for i in range(1, num_products + 1)],
    'aisle_id': np.random.randint(1, 20, num_products),
    'department_id': np.random.randint(1, 10, num_products)
}
df_products = pd.DataFrame(products_data)
print("\ndf_products (simulado):")
print(df_products.head())

# Simular a tabela 'aisles'
aisles_data = {
    'aisle_id': range(1, 20),
    'aisle': [f'Corredor_{i}' for i in range(1, 20)]
}
df_aisles = pd.DataFrame(aisles_data)
print("\ndf_aisles (simulado):")
print(df_aisles.head())

# Simular a tabela 'departments'
departments_data = {
    'department_id': range(1, 10),
    'department': [f'Departamento_{i}' for i in range(1, 10)]
}
df_departments = pd.DataFrame(departments_data)
print("\ndf_departments (simulado):")
print(df_departments.head())

# Simular 'order_products__prior' e 'order_products__train'
num_order_products_prior = 5000
num_order_products_train = 1000

order_products_prior_data = {
    'order_id': np.random.choice(df_orders[df_orders['eval_set'] == 'prior']['order_id'], num_order_products_prior),
    'product_id': np.random.randint(1, num_products + 1, num_order_products_prior),
    'add_to_cart_order': np.random.randint(1, 15, num_order_products_prior),
    'reordered': np.random.choice([0, 1], num_order_products_prior, p=[0.4, 0.6])
}
df_order_products_prior = pd.DataFrame(order_products_prior_data)

order_products_train_data = {
    'order_id': np.random.choice(df_orders[df_orders['eval_set'] == 'train']['order_id'], num_order_products_train),
    'product_id': np.random.randint(1, num_products + 1, num_order_products_train),
    'add_to_cart_order': np.random.randint(1, 15, num_order_products_train),
    'reordered': np.random.choice([0, 1], num_order_products_train, p=[0.3, 0.7])
}
df_order_products_train = pd.DataFrame(order_products_train_data)

print("\ndf_order_products_prior (simulado):")
print(df_order_products_prior.head())
print("\ndf_order_products_train (simulado):")
print(df_order_products_train.head())

# ----------------------------------------------------------------------------
# 2. Pré-processamento Base: Concatenação e Junção de Tabelas
# ----------------------------------------------------------------------------

print("\n--- 2. Pré-processamento Base ---")

# Concatenação de order_products__prior e __train
df_order_products_combined = pd.concat([df_order_products_prior, df_order_products_train], ignore_index=True)
print(f"\nTotal de itens de pedido combinados: {len(df_order_products_combined)}")

# Junção (Merge) de todas as tabelas
df_master = df_order_products_combined.merge(df_orders, on='order_id', how='left')
df_master = df_master.merge(df_products, on='product_id', how='left')
df_master = df_master.merge(df_aisles, on='aisle_id', how='left')
df_master = df_master.merge(df_departments, on='department_id', how='left')

print("\ndf_master (DataFrame combinado):")
print(df_master.head())
print(f"Dimensões do df_master: {df_master.shape}")

# Otimização de tipos de dados (conforme mencionado na apresentação)
for col in ['order_dow', 'order_hour_of_day', 'eval_set', 'aisle', 'department']:
    if col in df_master.columns:
        df_master[col] = df_master[col].astype('category')

print("\nTipos de dados otimizados para colunas categóricas.")
print(df_master.info())

# ----------------------------------------------------------------------------
# 3. Criação de Atributos Derivados (Engenharia de Features)
# ----------------------------------------------------------------------------

print("\n--- 3. Criação de Atributos Derivados ---")

# 1. is_first_order (booleano)
# Requer saber qual é o primeiro pedido de cada usuário
df_orders_min_order = df_orders.groupby('user_id')['order_number'].min().reset_index()
df_orders_min_order.rename(columns={'order_number': 'first_order_number'}, inplace=True)
df_master = df_master.merge(df_orders_min_order, on='user_id', how='left')
df_master['is_first_order'] = (df_master['order_number'] == df_master['first_order_number']).astype(int)
df_master.drop(columns=['first_order_number'], inplace=True)
print("\n'is_first_order' criado:")
print(df_master[['user_id', 'order_number', 'is_first_order']].drop_duplicates().head())

# 2. order_type (categórica)
df_master['order_type'] = np.where(df_master['is_first_order'] == 1, 'Primeira Compra', 'Recompra')
df_master['order_type'] = df_master['order_type'].astype('category')
print("\n'order_type' criado:")
print(df_master[['order_id', 'is_first_order', 'order_type']].drop_duplicates().head())


# 3. product_reordered_status (renomeação)
df_master.rename(columns={'reordered': 'product_reordered_status'}, inplace=True)
print("\n'reordered' renomeado para 'product_reordered_status':")
print(df_master[['product_id', 'product_reordered_status']].head())


# 4. order_size_category (categórica)
# Primeiro, calcular o tamanho do pedido
order_sizes = df_master.groupby('order_id').size().reset_index(name='order_total_items')
df_master = df_master.merge(order_sizes, on='order_id', how='left')

bins = [0, 5, 15, np.inf] # 0-5 itens, 6-15 itens, >15 itens
labels = ['Pequeno', 'Médio', 'Grande']
df_master['order_size_category'] = pd.cut(df_master['order_total_items'], bins=bins, labels=labels, right=True)
df_master['order_size_category'] = df_master['order_size_category'].astype('category')
print("\n'order_size_category' criado:")
print(df_master[['order_id', 'order_total_items', 'order_size_category']].drop_duplicates().head())


# 5. add_to_cart_order_group (categórica)
bins_cart = [0, 1, 5, 10, np.inf] # 1º item, 2-5º, 6-10º, >10º
labels_cart = ['1º Item', '2º-5º Itens', '6º-10º Itens', '>10º Itens']
df_master['add_to_cart_order_group'] = pd.cut(df_master['add_to_cart_order'], bins=bins_cart, labels=labels_cart, right=True)
df_master['add_to_cart_order_group'] = df_master['add_to_cart_order_group'].astype('category')
print("\n'add_to_cart_order_group' criado:")
print(df_master[['add_to_cart_order', 'add_to_cart_order_group']].head())


# 6. has_new_product_in_order / 7. percent_new_products (para cada usuário/pedido)
# Esta é mais complexa e requer um histórico de compras por usuário
print("\nCalculando 'has_new_product_in_order' e 'percent_new_products'...")

# Para cada usuário, criar uma lista de produtos únicos já comprados
user_product_history = {}
for user_id in df_master['user_id'].unique():
    user_orders = df_master[df_master['user_id'] == user_id].sort_values('order_number')
    bought_products = set()
    for order_id in user_orders['order_id'].unique():
        current_order_products = user_orders[user_orders['order_id'] == order_id]['product_id'].tolist()

        new_products_in_order = 0
        total_products_in_order = len(current_order_products)

        if total_products_in_order > 0:
            for prod_id in current_order_products:
                if prod_id not in bought_products:
                    new_products_in_order += 1

            df_master.loc[df_master['order_id'] == order_id, 'has_new_product_in_order'] = 1 if new_products_in_order > 0 else 0
            df_master.loc[df_master['order_id'] == order_id, 'percent_new_products'] = (new_products_in_order / total_products_in_order) * 100
        else:
            df_master.loc[df_master['order_id'] == order_id, 'has_new_product_in_order'] = 0
            df_master.loc[df_master['order_id'] == order_id, 'percent_new_products'] = 0

        bought_products.update(current_order_products)

# Preencher NaN que podem ocorrer para order_ids sem itens, ou se o user_id não estava na simulação (raro)
df_master['has_new_product_in_order'].fillna(0, inplace=True)
df_master['percent_new_products'].fillna(0, inplace=True)
df_master['has_new_product_in_order'] = df_master['has_new_product_in_order'].astype(int)

print("\n'has_new_product_in_order' e 'percent_new_products' criados (exemplo por order_id):")
print(df_master[['order_id', 'user_id', 'has_new_product_in_order', 'percent_new_products']].drop_duplicates(subset=['order_id']).head())


# 8. price_per_unit (simulado)
# Simular preços baseados em departamento
department_prices = {
    'Departamento_1': (1.0, 5.0),  # frozen
    'Departamento_2': (0.5, 3.0),  # other
    'Departamento_3': (2.0, 10.0), # bakery
    'Departamento_4': (0.8, 7.0),  # produce
    'Departamento_5': (5.0, 20.0), # alcohol
    'Departamento_6': (3.0, 15.0), # international
    'Departamento_7': (1.5, 8.0),  # beverages
    'Departamento_8': (4.0, 12.0), # pets
    'Departamento_9': (1.2, 6.0)   # dry goods pasta
}

def assign_simulated_price(row):
    dept = row['department']
    if dept in department_prices:
        low, high = department_prices[dept]
        return np.random.uniform(low, high)
    return np.random.uniform(0.5, 20.0) # Default para departamentos não mapeados

df_master['price_per_unit'] = df_master.apply(assign_simulated_price, axis=1)
print("\n'price_per_unit' (simulado) criado:")
print(df_master[['product_name', 'department', 'price_per_unit']].head())

# ----------------------------------------------------------------------------
# 4. Criação de Atributos Derivados Agregados (Exemplos para Análise 3)
# ----------------------------------------------------------------------------

print("\n--- 4. Criação de Atributos Derivados Agregados (por Usuário) ---")

# Calcular Total de Pedidos por Usuário
df_user_features = df_orders.groupby('user_id').agg(
    total_pedidos=('order_id', 'nunique')
).reset_index()

# Calcular Média de Itens por Pedido por Usuário
# Primeiro, contar itens por pedido
items_per_order = df_master.groupby('order_id').size().reset_index(name='num_items')
# Juntar com df_orders para ter o user_id
items_per_order = items_per_order.merge(df_orders[['order_id', 'user_id']], on='order_id', how='left')
# Média de itens por pedido para cada usuário
avg_items_per_user_order = items_per_order.groupby('user_id')['num_items'].mean().reset_index(name='media_itens_por_pedido')
df_user_features = df_user_features.merge(avg_items_per_user_order, on='user_id', how='left')


# Calcular Taxa de Recompra Percentual por Usuário
reorder_rate_per_user = df_master.groupby('user_id')['product_reordered_status'].mean().reset_index(name='taxa_recompra_percentual_raw')
reorder_rate_per_user['taxa_recompra_percentual'] = (reorder_rate_per_user['taxa_recompra_percentual_raw'] * 100).round(2)
df_user_features = df_user_features.merge(reorder_rate_per_user[['user_id', 'taxa_recompra_percentual']], on='user_id', how='left')

# Criar Perfil de Recompra (Categórica)
def classify_loyalty(rate):
    if rate >= 95:
        return 'Super Fiel'
    elif rate >= 80:
        return 'Fiel'
    elif rate >= 50:
        return 'Regular'
    else:
        return 'Ocasional'

df_user_features['perfil_recompra'] = df_user_features['taxa_recompra_percentual'].apply(classify_loyalty)
df_user_features['perfil_recompra'] = df_user_features['perfil_recompra'].astype('category')


print("\nDataFrame de Features Agregadas por Usuário:")
print(df_user_features.head())
print(df_user_features['perfil_recompra'].value_counts())

print("\n--- Demonstração de Atributos Derivados Concluída! ---")
print("O DataFrame 'df_master' agora contém a maioria dos atributos derivados por item de pedido.")
print("O DataFrame 'df_user_features' contém atributos derivados agregados por usuário.")
```
